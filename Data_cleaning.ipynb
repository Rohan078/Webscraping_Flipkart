```python
import pandas as pd

df = pd.read_csv("flipkart_data.csv")


df.columns = [col.strip().lower().replace(" ", "_") for col in df.columns]


df = df[df['actual_price'].notna()]
df = df[df['actual_price'].astype(str).str.lower() != 'n/a']


df['actual_price'] = df['actual_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False)
df = df[df['actual_price'] != ''] 
df['actual_price'] = df['actual_price'].astype(int)


df['offer_price'] = df['offer_price'].astype(str).str.replace('₹', '', regex=False).str.replace(',', '', regex=False).fillna('0').astype(int)


df['no.of_ratings'] = df['no.of_ratings'].astype(str).str.replace(',', '', regex=False)
df['no.of_ratings'] = pd.to_numeric(df['no.of_ratings'], errors='coerce').fillna(0).astype(int)


df['reviews'] = pd.to_numeric(df['reviews'].astype(str).str.replace(',', '', regex=False), errors='coerce').fillna(0).astype(int)

df['averager_rating'] = pd.to_numeric(df['averager_rating'], errors='coerce').fillna(0).astype(float)


df['offer_%'] = df['offer_%'].astype(str).str.replace('% off', '', regex=False).fillna('0').astype(int)


df.to_csv("flipkart_data_cleaned.csv", index=False)

```

```python
import pandas as pd

df = pd.read_csv("flipkart_data_cleaned.csv")

columns_to_keep = [
    'product_name', 'actual_price', 'offer_price',
    'no.of_ratings', 'reviews', 'averager_rating',
    'brand', 'offer_%'
]

df_reduced = df[columns_to_keep].head(500)

df_reduced.to_csv("flipkart_data_reduced.csv", index=False)

print("Reduced dataset saved to 'flipkart_data_reduced.csv'")
```
